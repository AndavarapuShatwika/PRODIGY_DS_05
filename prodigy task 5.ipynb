```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
data = pd.read_csv('accident_data .csv')
```

```python
#data preprocessing
# Check for missing values
missing_values = data.isnull().sum()
```

```python
# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Check data types
data.info()
```

```python
# Descriptive statistics
summary_stats = data.describe()
```

```python
# Visualize accident distribution by road conditions
plt.figure(figsize=(12, 6))
sns.countplot(x='Road Conditions', data=data)
plt.xticks(rotation=45)
plt.xlabel('Road Conditions')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Road Conditions')
plt.show()
```

```python
# Visualize accident distribution by weather conditions
plt.figure(figsize=(12, 6))
sns.countplot(x='Weather Conditions', data=data)
plt.xticks(rotation=45)
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Weather Conditions')
plt.show()
```

```python
#contributing factors analysis
plt.figure(figsize=(12, 6))
sns.countplot(x='Vehicle Speed', data=data)
plt.xticks(rotation=90)
plt.xlabel('Vehicle Speed')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Vehicle Speed')
plt.show()
```
